```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('D:\PRODIGY\customer_data.csv')
data.head()
```

```python
#Histogram for age
plt.figure(figsize=(8,7))
plt.hist(data['age'])
plt.xlabel('age')
plt.title('Customer purchased based on age')
plt.show()
```

```python
data.hist(figsize=(10,10))
plt.show()
```

```python
#bar plot
gender_counts=data['gender'].value_counts()
plt.figure(figsize=(8, 6))
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
gender_counts.plot(kind='bar', color=['blue', 'pink'], alpha=0.7)
plt.show()
```

```python
gender_counts=data['education'].value_counts()
plt.figure(figsize=(8, 6))
plt.title('Education')
plt.xlabel('educationn')
plt.ylabel('Count')
gender_counts.plot(kind='bar', color=['red','blue','green','black'], alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,7))
plt.hist(data['satisfaction_score'])
plt.xlabel('satisfaction score ')
plt.title('Customer satisfaction score ')
plt.show()
```

```python
gender_counts=data['purchase_frequency'].value_counts()
plt.figure(figsize=(8, 6))
plt.title('Purchase frequency of customer')
plt.xlabel('purchase_frequency ')
plt.ylabel('Count')
gender_counts.plot(kind='bar', color=['yellow','orange','green',], alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=data,x='region')
```
